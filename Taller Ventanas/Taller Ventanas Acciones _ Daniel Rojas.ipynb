```python
from pyspark.sql import SparkSession
from pyspark.sql import Window
from pyspark.sql import functions as f
from pyspark.sql.functions import col
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import os

os.environ["JAVA_HOME"]="/usr/lib/jvm/java-8-openjdk-amd64"

spark = SparkSession.builder.appName('ABC').getOrCreate();
```

```python
df_ec = spark.read.option('inferSchema','true').csv('EC*',header = True)
df_aval = spark.read.option('inferSchema','true').csv('AVAL*',header = True)
df_avh = spark.read.option('inferSchema','true').csv('AVH*',header = True)
```

```python
df_ec.printSchema();
df_aval.printSchema();
df_avh.printSchema();
```

## punto 1 Porcentaje Retorno Diario



```python
window1 = Window.orderBy(df_ec.Date); 
df_daily_EC= df_ec.select(f.col('Date').cast('date'),'Open','Close',((df_ec['Close']-df_ec['Open'])*100/df_ec['Open']).alias('Porcentaje %')).show();
df_daily_AVAL= df_aval.select(f.col('Date').cast('date'),'Open','Close',((df_aval['Close']-df_aval['Open'])*100/df_aval['Open']).alias('Porcentaje %')).show();
df_daily_AVH= df_avh.select(f.col('Date').cast('date'),'Open','Close',((df_avh['Close']-df_avh['Open'])*100/df_avh['Open']).alias('Porcentaje %')).show()
```

## punto 2 Porcentaje Retorno mensual

```python
window2 = Window.orderBy(df_ec.Open,df_ec.Close); 
df_month_EC= df_ec.select(f.year(f.col('Date').cast('date')).alias('year'),f.month(f.col('Date').cast('date')).alias('month'),'Open','Close');
df_month2_EC= df_month_EC.groupBy('year','Month').avg('Open','Close').sort('year','Month');
df_month3_EC= df_month2_EC.select('year','Month','avg(Open)','avg(Close)',((df_month2_EC['avg(Close)']-df_month2_EC['avg(Open)'])*100/df_month2_EC['avg(Open)']).alias('Porcentaje'));
df_month3_EC.show();

df_month_AVAL= df_aval.select(f.year(f.col('Date').cast('date')).alias('year'),f.month(f.col('Date').cast('date')).alias('month'),'Open','Close');
df_month2_AVAL= df_month_AVAL.groupBy('year','Month').avg('Open','Close').sort('year','Month');
df_month3_AVAL= df_month2_AVAL.select('year','Month','avg(Open)','avg(Close)',((df_month2_AVAL['avg(Close)']-df_month2_AVAL['avg(Open)'])*100/df_month2_AVAL['avg(Open)']).alias('Porcentaje'));
df_month3_AVAL.show();

df_month_AVH= df_avh.select(f.year(f.col('Date').cast('date')).alias('year'),f.month(f.col('Date').cast('date')).alias('month'),'Open','Close');
df_month2_AVH= df_month_AVH.groupBy('year','Month').avg('Open','Close').sort('year','Month');
df_month3_AVH= df_month2_AVH.select('year','Month','avg(Open)','avg(Close)',((df_month2_AVH['avg(Close)']-df_month2_AVH['avg(Open)'])*100/df_month2_AVH['avg(Open)']).alias('Porcentaje'));
df_month3_AVH.show();
```

## Punto 3 MES con mayor Porcentage de Retorno


```python
df_month_max_EC = df_month3_EC.select('year','Month','Porcentaje').groupBy('year').max('Month','Porcentaje').show()
df_month_max_AVAL = df_month3_AVAL.select('year','Month','Porcentaje').groupBy('year').max('Month','Porcentaje').show()
df_month_max_AVH = df_month3_AVH.select('year','Month','Porcentaje').groupBy('year').max('Month','Porcentaje').show()
```

## Punto 4 Media Movil 20 DIAS ECOPETROL --AVAL

## Punto 5 Grafica Comportamiento 3 Acciones

```python

window1 = Window.orderBy(df_ec.Date);


fig, ax = plt.subplots();
EC_pd = df_ec.toPandas();
AVAL_pd = df_aval.toPandas();
AVH_pd = df_avh.toPandas();
EC_pd.set_index("Date");
AVAL_pd.set_index("Date");
AVH_pd.set_index("Date");

ax.plot(EC_pd.Date, EC_pd.Close, label="Ecopetrol")
ax.plot(AVAL_pd.Date, AVAL_pd.Close, label="AVAL")
ax.plot(AVAL_pd.Date, AVH_pd.Close, label="AVH")
ax.legend()
```

## Punto 6 Graficas de BOLLIGER K = 2 (Ecopetrol)

```python

window1 = Window.orderBy(df_ec.Date).rowsBetween(-20,0)
df17 = df_ec.select('Date', f.avg('Close').over(window1).alias('Average'))
df18 = df_ec.select('Date',(f.avg('Close').over(window1).alias('Average') + f.sqrt(f.pow(f.abs(f.col('Close') - (f.avg('Close').over(window1).alias('Average'))),2)/f.count('Date').over(window1))*2).alias('movilSup'));
df19 = df_ec.select('Date',(f.avg('Close').over(window1).alias('Average') - f.sqrt(f.pow(f.abs(f.col('Close') - (f.avg('Close').over(window1).alias('Average'))),2)/f.count('Date').over(window1))*2).alias('movilInf'));
fig, ax = plt.subplots()
EC_pd = df_ec.toPandas()
EC_AVG = df17.toPandas()
EC_sup  = df18.toPandas();
EC_inf = df19.toPandas();
EC_pd.set_index("Date");
EC_AVG.set_index("Date");
EC_sup.set_index("Date");
EC_inf.set_index("Date");
ax.plot(EC_AVG.Date, EC_AVG.Average, label="Mobile Half")
ax.plot(EC_sup.Date, EC_sup.movilSup, label="Mobile Sup")
ax.plot(EC_inf.Date, EC_inf.movilInf, label="Mobile Inf")
ax.plot(EC_pd.Date, EC_pd.Close, label="Real Value");
ax.legend()
```
